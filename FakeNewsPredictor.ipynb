```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
#from sklearn.svm import SVC
#from sklearn.naive_bayes import MultinomialNB
#from sklearn.neighbors import KNeighborsClassifier


true = pd.read_csv('True.csv')
fake = pd.read_csv('Fake.csv')

true.head()


```

```python
fake.head()
```

```python
true['target'] = 1
fake['target'] = 0
true.head()
```

```python
news = pd.concat([true, fake]).sample(frac=1).reset_index(drop = True)
news.head()
```

```python
news = news.drop(['title','subject', 'date'], axis = 1)
news.head()
```

```python
import re
import string

def clean_text(text):
    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', ' ', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove numbers
    text = re.sub(r'\d+', ' ', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text


```

```python
news['text'] = news['text'].apply(clean_text)
news.head()
```

```python
x = news['text']
y = news['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
v =  TfidfVectorizer()
```

```python
xv_train = v.fit_transform(x_train)
xv_test = v.transform(x_test)
```

```python
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))

```

```python
print(confusion_matrix(y_test, pred_lr))
```

```python
print(accuracy_score(y_test, pred_lr))
```

```python
DTC = DecisionTreeClassifier()
DTC.fit(xv_train, y_train)
```

```python
pred_dt = DTC.predict(xv_test)
DTC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
rfc = RandomForestClassifier()
rfc.fit(xv_train, y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
print(confusion_matrix(y_test, pred_rfc))
```

```python
print(accuracy_score(y_test, pred_rfc))
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(xv_train, y_train)
```

```python
pred_gbc = gbc.predict(xv_test)
gbc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
print(confusion_matrix(y_test, pred_gbc))
```

```python
print(accuracy_score(y_test, pred_gbc))
```

```python
def output_label(n):
    return "Not A Fake News" if n == 1 else "Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(clean_text)

    new_x_test = new_def_test["text"]
    new_xv_test = v.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
  #  pred_DT = DTC.predict(new_xv_test)
    pred_RFC = rfc.predict(new_xv_test)
    pred_GBC = gbc.predict(new_xv_test)

    print("\n===== PREDICTIONS =====")
    print("Logistic Regression (LR):", output_label(pred_LR[0]))
    print("Gradient Boosting Classifier (GBC):", output_label(pred_GBC[0]))
    print("Random Forest Classifier (RFC):", output_label(pred_RFC[0]))

```
